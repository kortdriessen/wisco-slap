```python
# this snippet of code shows how to get all of the currently available hypnograms that are ready to go for training.
import wisco_slap as wis
import os
macq = wis.meta.get.acq_master()
final_hypno_paths = []
for subject in macq:
    for exp in macq[subject]:
        for acq_id in macq[subject][exp]:
            try:
                loc = acq_id.split('--')[0]
                acq = acq_id.split('--')[1]
                sb = wis.meta.sync.get_acq_sync_block(subject, exp, loc, acq)
                hypno_dir = os.path.join(wis.defs.anmat_root, subject, exp, 'sync_block_data', f'sync_block-{sb}', 'hypnograms', 'training_labels')
                if os.path.exists(hypno_dir):
                    hypno_files = os.listdir(hypno_dir)
                    assert len(hypno_files) == 1, f"Expected exactly one hypno file in {hypno_dir}, found {len(hypno_files)}"
                    hypno_path = os.path.join(hypno_dir, hypno_files[0])
                    if hypno_path not in final_hypno_paths:
                        final_hypno_paths.append(hypno_path)
                else:
                    continue
            except Exception as e:
                continue
final_hypno_paths
```
